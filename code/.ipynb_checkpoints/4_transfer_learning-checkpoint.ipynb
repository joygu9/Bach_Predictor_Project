## Transfer Learning

```python
# using https://adapt-python.github.io/adapt/generated/adapt.instance_based.TrAdaBoostR2.html
from adapt.instance_based import TrAdaBoostR2
from sklearn.linear_model import Ridge, Lasso

#load in bachelor dataset, this will be the source dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from utils import *

df_bachelor = pd.read_csv("data_processed/bachelor_preprocessed.csv")
X,y = encode_data(df_bachelor)

# Split the training, validation, test data
np.random.seed(100)

#default is trainsize=0.7, valsize=0.1, testsize=0.2
X_train, y_train, X_val, y_val, X_test, y_test = split_data(X,y)

# Combine training and validation into a new training set for final test performance
y_train_combined = pd.concat([y_train, y_val])
X_train_combined  = pd.concat([X_train, X_val])

#standardize values based on training statistics
X_test = standardize_data(X_train_combined, X_test)
X_train_combined = standardize_data(X_train_combined, X_train_combined)

X_test_S = X_test
X_train_combined_S = X_train_combined
y_test_S = y_test
y_train_combined_S = y_train_combined
X_val_S = X_val
y_val_S = y_val
X_train_S = X_train
y_train_S = y_train
X_S = X
```

```python
#Now load the bachelorette data - this is our target dataset
df_bachelorette = pd.read_csv("data_processed/bachelorette_preprocessed.csv")
X,y = encode_data(df_bachelorette)
print(len(X))

# Split the training, validation, test data
np.random.seed(15)

#default is trainsize=0.7, valsize=0.1, testsize=0.2
X_train, y_train, X_val, y_val, X_test, y_test = split_data(X,y)
print(len(X_train), len(y_train))
print(len(X_val), len(y_val))
print(len(X_test), len(y_test))
print("Winners in training: ", sum(y_train == 11))
print("Winners in validation: ", sum(y_val == 11))
print("Winners in test: ", sum(y_test == 11))


# Combine training and validation into a new training set for final test performance
y_train_combined = pd.concat([y_train, y_val])
X_train_combined  = pd.concat([X_train, X_val])

#standardize values based on training statistics
X_test = standardize_data(X_train_combined, X_test)
X_train_combined = standardize_data(X_train_combined, X_train_combined)

X_test_T = X_test
X_train_combined_T = X_train_combined
y_test_T = y_test
y_train_combined_T = y_train_combined
X_val_T = X_val
y_val_T = y_val
X_train_T = X_train
y_train_T = y_train
X_T = X

```

### Transfer learning training and test

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# TL - training on Source
#use validation data to select model parameters
lambda_L1 = 2**(-4.5)
lambda_L2 = 2**(1.5)
best_mse = 1000
best_n = 0
best_rate = 0
for n_est in [1, 5, 10, 15, 20]:
    for rate in [0.1,0.5, 1, 2]:
        model_TF = TrAdaBoostR2(Ridge(alpha=lambda_L2, max_iter=1000000,fit_intercept=False),
                                n_estimators=n_est,lr=rate, Xt=X_train_T, yt=y_train_T, random_state=0)
        model_TF.fit(X_train_S, y_train_S)
        y_pred = model_TF.predict_estimator(X_val_T)
        val_mse = mean_squared_error(y_val_T, y_pred)
        print("Val MSE: %.3f" %val_mse)
        
        if val_mse < best_mse:
            best_mse = val_mse
            best_n = n_est
            best_rate = rate
```

```python
print(best_n, best_rate, best_mse)
```

```python
# performance on test set
# from sklearn.metrics import r2_score
model_TF = TrAdaBoostR2(Ridge(alpha=lambda_L2, max_iter=1000000,fit_intercept=False),
                                n_estimators=n_est,lr=rate, Xt=X_train_T, yt=y_train_T, random_state=0)

model_TF.fit(X_train_combined_S, y_train_combined_S)
y_pred = model_TF.predict_estimator(X_test_T)
y_pred_train = model_TF.predict(X_train_combined_S)
test_mse = mean_squared_error(y_test_T, y_pred)
print("Transfer Learning: TrAdaBoostR2 with Ridge")
print_metrics(y_pred, y_test_T, y_pred_train, y_train_combined_S)
print(model_TF.predict_weights("target"))
print("R^2: ", model_TF.score(X_test_T, y_test_T))
plot_scatter_pred_true(y_pred, y_test_T, "TrAdaBoostR2 with Ridge (L2) Regression")

```
