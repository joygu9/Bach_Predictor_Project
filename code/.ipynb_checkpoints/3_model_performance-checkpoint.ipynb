# Model Performance for Test Data
### Retrain the best models using training and validation data, calculate MSE on test data

```python
#load in bachelor dataset, do one-hot encoding for hometown, and standardize Age, NumRoses3, FirstDate
import pandas as pd
import numpy as np
from utils import *
import matplotlib.pyplot as plt


df_bachelor = pd.read_csv("data_processed/bachelor_preprocessed.csv")
X,y = encode_data(df_bachelor)

# Split the training, validation, test data
np.random.seed(100)

#default is trainsize=0.7, valsize=0.1, testsize=0.2
X_train, y_train, X_val, y_val, X_test, y_test = split_data(X,y)
print(len(X_train), len(y_train))
print(len(X_val), len(y_val))
print(len(X_test), len(y_test))
# Combine training and validation into a new training set for final test performance
y_train_combined = pd.concat([y_train, y_val])
X_train_combined  = pd.concat([X_train, X_val])
print(len(X_train_combined), len(y_train_combined))

#standardize values based on training statistics
X_test = standardize_data(X_train_combined, X_test)
X_train_combined = standardize_data(X_train_combined, X_train_combined)
print(len(X_train_combined), len(y_train_combined))

print("Winners in training: ", sum(y_train == 11))
print("Winners in validation: ", sum(y_val == 11))
print("Winners in test: ", sum(y_test == 11))

```

```python
#calculate e_M for generalization bound\
M=1
N = len(y_test)
delta = 0.05
e_M = np.sqrt((1/(2*N))*np.log(2*M/delta))
print("Number of test samples: ", N)
print("e_M = ", e_M)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Retrain models on this combined dataset

#Linear Reg
reg_linear = LinearRegression(fit_intercept=False)
reg_linear = reg_linear.fit(X_train_combined, y_train_combined)
y_pred = reg_linear.predict(X_test)
y_pred_train = reg_linear.predict(X_train_combined)
print("Linear Regression")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print(reg_linear.coef_)
print("R^2: ", reg_linear.score(X_test, y_test))
plot_scatter_pred_true(y_pred, y_test, "Linear Regression")

#Lasso
lambda_L1 = 2**(-4.5)
reg_L1 = Lasso(alpha=lambda_L1, max_iter=1000000,fit_intercept=False)
reg_L1 = reg_L1.fit(X_train_combined, y_train_combined)
y_pred = reg_L1.predict(X_test)
y_pred_train = reg_L1.predict(X_train_combined)
print("Lasso (L1) Regression")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print(reg_L1.coef_)
print("R^2: ", reg_L1.score(X_test, y_test))
plot_scatter_pred_true(y_pred, y_test, "Lasso (L1) Regression")


#Ridge
lambda_L2 = 2**(1.5)
reg_L2 = Ridge(alpha=lambda_L2, max_iter=1000000,fit_intercept=False)
reg_L2 = reg_L2.fit(X_train_combined, y_train_combined)
y_pred = reg_L2.predict(X_test)
y_pred_train = reg_L2.predict(X_train_combined)
print("Ridge (L2) Regression")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print(reg_L2.coef_)
print("R^2: ", reg_L2.score(X_test, y_test))
plot_scatter_pred_true(y_pred, y_test, "Ridge (L2) Regression")

#Random Forest
n_tree = 50
max_d = 2
reg_RF = RandomForestRegressor(n_estimators=n_tree, max_depth=max_d, random_state=100)
reg_RF.fit(X_train_combined, y_train_combined)
y_pred = reg_RF.predict(X_test)
y_pred_train = reg_RF.predict(X_train_combined)
print("Random Forest Regressor")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print(reg_RF.feature_importances_)
print("R^2: ", reg_RF.score(X_test, y_test))
plot_scatter_pred_true(y_pred, y_test, "Random Forest Regressor")
```

```python
# PCA
from sklearn.decomposition import PCA

# Split the training, validation, test data
np.random.seed(100)

#default is trainsize=0.7, valsize=0.1, testsize=0.2
X_train, y_train, X_val, y_val, X_test, y_test = split_data(X,y)

# Combine training and validation into a new training set for final test performance
y_train_combined = pd.concat([y_train, y_val])
X_train_combined  = pd.concat([pd.DataFrame(X_train), pd.DataFrame(X_val)])

#standardize values based on training statistics
X_test = standardize_data(X_train_combined, X_test)
X_train_combined = standardize_data(X_train_combined, X_train_combined)


pca = PCA(n_components=3)
pca.fit(X_train_combined)
print("explained_variance",pca.explained_variance_ratio_)

X_test = pca.transform(X_test)
X_train_combined = pca.transform(X_train_combined)


n_tree = 10
max_d = 2
reg_RF_pca = RandomForestRegressor(n_estimators=n_tree, max_depth=max_d, random_state=100)
reg_RF_pca.fit(X_train_combined, y_train_combined)
y_pred = reg_RF_pca.predict(X_test)
y_pred_train = reg_RF_pca.predict(X_train_combined)
print("Random Forest Regressor - PCA")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print("R^2: ", reg_RF_pca.score(X_test, y_test))
print(reg_RF_pca.feature_importances_)
plot_scatter_pred_true(y_pred, y_test, "PCA with Random Forest Regressor")
```

### Other Baseline systems

```python
from sklearn.metrics import r2_score
#Trivial Baseline systems
# Split the training, validation, test data
np.random.seed(100)
#default is trainsize=0.7, valsize=0.1, testsize=0.2
X_train, y_train, X_val, y_val, X_test, y_test = split_data(X,y)
# Combine training and validation into a new training set for final test performance
y_train_combined = pd.concat([y_train, y_val])
X_train_combined  = pd.concat([X_train, X_val])



#random of y between 1 and 11
np.random.seed(500)
y_pred = np.random.randint(1,12, len(y_test))

print("Random ElimWeek")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print("R^2: ", r2_score(y_test, y_pred))
plot_scatter_pred_true(y_pred, y_test, "Trivial Baseline: Random")

# Nontrivial Baseline
#using simple linear regression on age
#Linear Reg
X_age_train_combined = X_train_combined.iloc[:,0].values.reshape(-1, 1)
X_age_test = X_test.iloc[:,0].values.reshape(-1, 1)
reg_linear = LinearRegression(fit_intercept=True)
reg_linear = reg_linear.fit(X_age_train_combined, y_train_combined)
y_pred = reg_linear.predict(X_age_test)
y_pred_train = reg_linear.predict(X_age_train_combined)
print("Linear Regression")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print("R^2: ", reg_linear.score(X_age_test, y_test))
print(reg_linear.coef_)
plot_scatter_pred_true(y_pred, y_test, "Nontrivial Baseline: Linear Regression on Age")


plt.clf()
plt.scatter(X_age_train_combined, y_train_combined)
plt.scatter(X_age_test, y_test)
plt.scatter(X_age_test, y_pred)
plt.xlabel("Age")
plt.ylabel("elimination week")
plt.title("Trivial Baseline (Linear Regression with Age) Test Prediction")
plt.legend(["training data","test data", "prediction"])
plt.show()

from sklearn.preprocessing import PolynomialFeatures
#quadratic with age
poly = PolynomialFeatures(degree=2, include_bias=False)
X_age_train_combined_2 = poly.fit_transform(X_age_train_combined)
X_age_test_2 = poly.fit_transform(X_age_test)

reg_poly = LinearRegression(fit_intercept=False)
reg_poly = reg_poly.fit(X_age_train_combined_2, y_train_combined)
y_pred = reg_poly.predict(X_age_test_2)
y_pred_train = reg_poly.predict(X_age_train_combined_2)
print("Polynomial Regression Degree 2")
print_metrics(y_pred, y_test, y_pred_train, y_train_combined, e_M)
print("R^2: ", reg_poly.score(X_age_test_2, y_test))
print(reg_poly.coef_)
plot_scatter_pred_true(y_pred, y_test,
                       "Nontrivial Baseline: Deg 2 Polynomial Regression on Age")


plt.clf()
plt.scatter(X_age_train_combined, y_train_combined)
plt.scatter(X_age_test, y_test)
plt.scatter(X_age_test, y_pred)
plt.xlabel("Age")
plt.ylabel("elimination week")
plt.title("Trivial Baseline (Deg2 Polynomial Regression with Age) Test Prediction")
plt.legend(["training data","test data", "prediction"])
plt.show()
```
