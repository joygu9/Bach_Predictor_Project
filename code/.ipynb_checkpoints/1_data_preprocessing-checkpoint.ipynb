### Join the bachelor contestant data

```python
#split the 538 data
data_538_bachelor = data_538[data_538["SHOW"] == "Bachelor"]
data_538_bachelorette = data_538[data_538["SHOW"] == "Bachelorette"]


#reference code from https://github.com/KwameTaylor/bachelorette-predictor/blob/main/wrangle.py

# Define function to create string for contestant id so we can join the dataset
def create_contestant_ID(df):
    #make a copy of the dataframe
    df_new = df.copy()
    
    # Split to first name, last initial, then join with season
    df_new['first_name'] = df.Name.replace(to_replace = '( )\w+.', value = '', regex = True)
    df_new['last_initial'] = df_new['Name'].str.extract(r'( \w{1})')
    df_new.last_initial = df_new.last_initial.fillna('X')
    df_new['CONTESTANT'] = df_new['Season'].astype(str)
    df_new['CONTESTANT'] = df_new['CONTESTANT'].str.zfill(2) + '_'+ df_new['first_name'].str.upper() + '_' + df_new['last_initial'].str.upper()
    df_new['CONTESTANT'] = df_new['CONTESTANT'].str.replace(" ", "")
    return df_new


#get rid of duplicate ids
data_bachelor_new = create_contestant_ID(data_bachelor)
data_bachelor_new = data_bachelor_new.drop_duplicates(subset="CONTESTANT")
print("length of bachelor contestants with unique ids: ", len(data_bachelor_new))

data_bachelor_joined = data_bachelor_new.join(data_538_bachelor.set_index('CONTESTANT'), on='CONTESTANT', how="inner")
print("length of joined dataset: ", len(data_bachelor_joined))

#check how well the join worked, then workshop some fixes in case names aren't right
for season_num in range(1,22):
    data_test = data_bachelor_new[data_bachelor_new["Season"] == season_num]
    data_test2 = data_bachelor_joined[data_bachelor_joined["SEASON"] == str(season_num).zfill(2)]
    print(season_num, len(data_test), len(data_test2))
```

### Prepare columns for machine learning, drop unnecessary columms
#### replace Nans for ElimWeek - which is our y column

```python
# Replace NaNs for Elimination Week
data_bachelor_joined[np.isnan(data_bachelor_joined["ElimWeek"])]

# Rachel and Raven were eliminated, everyone else was the winner of their season
# let week 11 be the elimination week for winners, even though some seasons didn't get to that point
data_nan = data_bachelor_joined.loc[np.isnan(data_bachelor_joined["ElimWeek"])]
data_nan.iloc[:,8:21]
```

```python
# Fill in the nans with the elimination week

def fill_elimweek(df):
    col = np.zeros(len(df))
    for i in range(len(df)):
        elim_info = df.iloc[i, 11:21]
        for i_e in range(len(elim_info)):
            if elim_info[i_e] == 'W':
                col[i] = 11
            if elim_info[i_e] == 'E':
                col[i] = i_e+1
    return col

data_nan = data_bachelor_joined.loc[np.isnan(data_bachelor_joined["ElimWeek"])]
new_elim_week = fill_elimweek(data_nan)
data_bachelor_joined.loc[np.isnan(data_bachelor_joined["ElimWeek"]), "ElimWeek"] = new_elim_week
data_bachelor_joined.head()
```

#### Create First Impression Rose (FIR) column: 1=got the FIR, 0=did not get the FIR

```python
def fill_FIR(df):
    FIR = np.zeros(len(df))
    for i in range(len(df)):
        elim_info = df.iloc[i, 11] #only need to look at elimination week 1
        if elim_info == 'R1':
            FIR[i] = 1
    return FIR

data_bachelor_joined["FIR"] = fill_FIR(data_bachelor_joined)
```

#### Create column for number of roses received in first 3 weeks (usually group date roses)

```python
def fill_NumRoses3(df):
    numroses = np.zeros(len(df))
    for i in range(len(df)):
        elim_info = df.iloc[i, 11:14] #only need to look at elimination week 1-3
        for i_e in range(len(elim_info)):
            if elim_info[i_e] == 'R1' or elim_info[i_e] == 'R' :
                numroses[i] = numroses[i] + 1
    return numroses

data_bachelor_joined["NumRoses3"] = fill_NumRoses3(data_bachelor_joined)
```

#### Create Week of First 1-1 Date Column

```python
def fill_FirstDate(df):
    firstdate = np.zeros(len(df))
    for i in range(len(df)):
        date_info = np.array(df.iloc[i, 21:31])
        inds = np.where(date_info == "D1")
        if len(inds[0]) != 0:
            firstdate[i] = inds[0][0] + 1 #week of first 1-1 date
    return firstdate

data_bachelor_joined["FirstDate"] = fill_FirstDate(data_bachelor_joined)

#remove unnecessary columns
data_bachelor_joined_dropped = data_bachelor_joined.drop(columns=["Name","first_name", "last_initial", "SHOW", "SEASON", "Season","DATES-1", 
                                   "DATES-2","DATES-3","DATES-4","DATES-5","DATES-6","DATES-7","DATES-8","DATES-9", "DATES-10",
                                   "ELIMINATION-1" ,"ELIMINATION-2","ELIMINATION-3" ,"ELIMINATION-4" ,"ELIMINATION-5" ,"ELIMINATION-6" ,
                                    "ELIMINATION-7" ,"ELIMINATION-8" ,"ELIMINATION-9" ,"ELIMINATION-10" ])
data_bachelor_joined_dropped = data_bachelor_joined_dropped.set_index("CONTESTANT")
data_bachelor_joined_dropped.head()
```
